# EDA - Visualisation: Scatterplot with Regression Line

## Load Modules

```python
import numpy as np
import seaborn as sns
```

## Load Data from Seaborn

```python
df = sns.load_dataset("penguins")
```

## Simple Scatterplot

```python
df.plot(x = "bill_length_mm", y = "flipper_length_mm", kind = "scatter")
```

## Scatterplot with regression line

```python
sns.lmplot(data=df, x="bill_length_mm", y="flipper_length_mm")
```

## Scatterplot with regression line per species

```python
sns.lmplot(data=df, x="bill_length_mm", y="flipper_length_mm", hue = "species")
```
