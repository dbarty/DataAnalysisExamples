# Outlier Detection and Correlation Analysis

## Generate example data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# create example data
np.random.seed(42)
data = {
    'age': np.random.normal(35, 10, 100),
    'salery': np.random.normal(50000, 15000, 100),
    'workhours_week': np.random.normal(40, 5, 100)
}

# add outliers
data['salery'][98] = 150000  # großer Ausreißer im Einkommen
data['salery'][99] = 200000  # weiterer großer Ausreißer
df = pd.DataFrame(data)
```

## 1. Identify outliers

### Create a Boxplot aka Box-Whisker-Plot
see Wikipedia [DE](https://de.wikipedia.org/wiki/Box-Plot) [EN](https://en.wikipedia.org/wiki/Box_plot)

```python
# Boxplot zur Identifikation von Ausreißern
plt.figure(figsize=(10,6))
sns.boxplot(data=df['salery'])
plt.title('Boxplot to identify outliers (salery)')
plt.show()
```

## 2. Analyse the correlation with Pearson's R

```python
# Calculate Pearson-Correlation
correlation_matrix = df.corr(method='pearson')
correlation_matrix
```

## 3. Create scatterplotfor better visualisation with correlation between salery and workhours_week

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['salery'], y=df['workhours_week'])
plt.title('Scatterplot: Salery vs. Workhours per Week')
plt.show()
```
