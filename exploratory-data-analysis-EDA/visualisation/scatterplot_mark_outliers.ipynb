# Mark outliers in Scatterplot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# create example data
np.random.seed(42)
data = {
    'age': np.random.normal(35, 10, 100),
    'salery': np.random.normal(50000, 15000, 100),
    'workhours_week': np.random.normal(40, 5, 100)
}

# add outliers
data['salery'][98] = 150000  # großer Ausreißer im Einkommen
data['salery'][99] = 200000  # weiterer großer Ausreißer
df = pd.DataFrame(data)
```

## Create scatterplot with correlation between salery and workhours_week

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['salery'], y=df['workhours_week'])
plt.title('Scatterplot: Salery vs. Workhours per Week')
plt.show()
```

## Create scatterplot with with marked outliers

```python


plt.figure(figsize=(8,5))
sns.scatterplot(x=df['salery'], y=df['workhours_week'])
plt.title('Scatterplot: Salery vs. Workhours per Week')

# find and mark the salery outliers >= 100k
outliers_salery_gt_100k = df[df['salery'] >= 100000]
plt.scatter(outliers_salery_gt_100k['salery'],outliers_salery_gt_100k['workhours_week'], 
            color='green', s=100, label='Outliers (Salery >= 100k)')

plt.legend()
plt.show()
```
