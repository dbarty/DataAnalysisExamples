# EDA - Visualisation: Scatterplot with Regression Line

## Load Modules

```python
import numpy as np
import seaborn as sns
```

## Load Data from Seaborn

```python
df = sns.load_dataset("penguins")
```

```python
def print_joint_plot(field):
    jp = sns.jointplot(data=df, x="bill_length_mm", y="flipper_length_mm", hue = field)
    items = df[field].unique()
    
    pal = sns.color_palette(n_colors=len(items))
    for item, color in zip(items, pal):
        sns.regplot(data=df[df[field] == item], x="bill_length_mm", y="flipper_length_mm", ax=jp.ax_joint)

print_joint_plot("species")
print_joint_plot("island")
```
