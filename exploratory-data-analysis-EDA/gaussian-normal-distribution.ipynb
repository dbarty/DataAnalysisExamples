# Exploratory Data Analysis (EDA) - Gaussian (normal) Distribution

## Load Modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

## Prepare Data

```python
# create example data (simulate normal distribution)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, standard deviation=10, 1000 values
df = pd.DataFrame(data, columns=['values'])
```

## Draw histplot for normal distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['values'], kde=True, color='blue', stat='density', label='Density', bins=50)
```

## Calculate values for mean, mode, quartiles and quantiles

```python
# Berechnungen
mean = df['values'].mean()
mode = df['values'].mode()[0]
quartiles = df['values'].quantile([0.25, 0.5, 0.75])
quantiles = df['values'].quantile([0.1, 0.9])
```

## Draw histplot with mean and mode

```python
plt.figure(figsize=(10, 6))
plt.title('Normal Distribution with Mean and Modus')

sns.histplot(df['values'], kde=True, color='blue', stat='density', label='Density', bins=50)

plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(mode, color='green', linestyle='--', label=f'Modus: {mode:.2f}')

plt.legend()
```

## Draw histplot with Quartiles

```python
plt.figure(figsize=(10, 6))
plt.title('Normal Distribution with Quartiles')
sns.histplot(df['values'], kde=True, color='blue', stat='density', label='Density', bins=50)

plt.axvline(quartiles[0.25], color='green', linestyle='--', label=f'1. Quartil: {quartiles[0.25]:.2f}')
plt.axvline(quartiles[0.5], color='orange', linestyle='--', label=f'Median: {quartiles[0.5]:.2f}')
plt.axvline(quartiles[0.75], color='green', linestyle='--', label=f'3. Quartil: {quartiles[0.75]:.2f}')

plt.legend()
```

## Draw histplot with Quantiles

```python
plt.figure(figsize=(10, 6))
plt.title('Normal Distribution with Quantiles')
sns.histplot(df['values'], kde=True, color='blue', stat='density', label='Density', bins=50)

plt.axvline(quantiles[0.1], color='green', linestyle='--', label=f'10%-Quantil: {quantiles[0.1]:.2f}')
plt.axvline(quantiles[0.9], color='green', linestyle='--', label=f'90%-Quantil: {quantiles[0.9]:.2f}')

plt.legend()
```
