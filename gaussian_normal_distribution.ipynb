# Gaussian (Normal) Distribution
See Wikipedia [DE](https://de.wikipedia.org/wiki/Normalverteilung) | [EN](https://en.wikipedia.org/wiki/Normal_distribution)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Simulierte Normalverteilung
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Darstellung der Normalverteilung
plt.figure(figsize=(10,6))
sns.histplot(normal_data, kde=True)
plt.title('Gaussian (Normal) Distribution')
plt.axvline(normal_data.mean(), color='green', linestyle='dashed', linewidth=2, label="Mean")
plt.legend()
plt.show()
```
