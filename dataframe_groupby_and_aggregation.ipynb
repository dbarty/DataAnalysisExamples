# Pandas: DataFrame groupby and aggregation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("penguins")
df.head()
```

## Create group objects

```python
# single field/index
grouped1 = df[["species", "bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]].groupby("species")
print(grouped1)

# multiple fields/indicies
grouped2 = df[["species", "island", "bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]].groupby(["species", "island"])
print(grouped2)
```

## Use aggregations (min, max, mean...)

```python
grouped1.agg(["min", "max", "mean"])
```

```python
grouped2.agg(["min", "max", "mean"])
```
